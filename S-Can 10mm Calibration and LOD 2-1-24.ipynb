# Calibration Curve

```python
additions=  pd.read_excel("/Users/beneck/Library/CloudStorage/OneDrive-NortheasternUniversity/Boise Project/SCAN Calibration/CalibrationAdditions.xlsx")
additions['Concentration (uM)']
```

```python
ten_mm_calibration['Concentration (uM)'] = additions['Concentration (uM)']
ten_mm_calibration
```

```python
linear_range = ten_mm_calibration.loc[0:20]
```

```python
coeff = np.polyfit(linear_range['220.00'], linear_range['Concentration (uM)'], 1)
vals = np.polyval(coeff, ten_mm_calibration['220.00'])
r2 = np.corrcoef(linear_range['220.00'], linear_range['Concentration (uM)'])[0,1]**2
coeff = np.append(coeff, r2)
vals
```

```python
fig, ax = plt.subplots(nrows = 1, ncols=2, figsize = (11, 6))

ten_mm_calibration.plot(x='220.00', y='Concentration (uM)', ax=ax[0], kind='scatter', ylabel= 'Nitrate Concentration (uM)', xlabel='Absorption at 220nm', xlim = (0,70), ylim = (0,500), title= "0-400 uM Nitrate")

ax[0].plot(ten_mm_calibration['220.00'], vals, color='b')
ax[0].text(5,400, 'Slope: %.2f' % (coeff[0]))
ax[0].text(5, 375, 'Intercept: %.2f' % (coeff[1]))
ax[0].text(5, 350, 'R2: %.4f' % (r2))

ten_mm_calibration.loc[0:12].plot(x='220.00', y='Concentration (uM)',kind='scatter', ax=ax[1], ylabel= 'Nitrate Concentration (uM)', xlabel='Absorption at 220nm', title= "0-30 uM Nitrate")
ax[1].plot(ten_mm_calibration.loc[0:12]['220.00'], vals[0:13], color='b')
#ax[1].text(5,300, 'Slope: %.2f' % (coeff[0]))
#ax[1].text(5, 275, 'Intercept: %.2f' % (coeff[1]))
#ax[1].text(5, 250, 'R2: %.4f' % (r2))
```
